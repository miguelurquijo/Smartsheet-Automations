```python
from IPython.display import display_html
import pandas as pd
import smartsheet
import logging
import time
import datetime
import os
import numpy as np
from datetime import datetime
smartsheet_client = smartsheet.Smartsheet('TOKEN HERE') #Smartsheet token
print("connected to Smartsheets as Smartadmin")
```

# Get Budget Sheet in a Data Frame

```python
display_html('SPREADSHEET PUBLISHED URL HERE', raw=True)
budgeturl = pd.read_html('SPREADSHEET PUBLISHED URL HERE') #google sheet URL
Budget_df = budgeturl[0]
new_header = Budget_df.iloc[0] #grab the first row for the header
Budget_df = Budget_df[1:] #take the data less the header row
Budget_df.columns = new_header #set the header row as the df header
Budget_df.dropna(axis = 0, inplace = True, how = 'all')
Budget_df = Budget_df[Budget_df['Growth ID'] != '']
Budget_df = Budget_df[Budget_df['Project'] != 'Missing URL']
Budget_df = Budget_df.replace(np.nan, '', regex=True)
Budget_df.columns
```
