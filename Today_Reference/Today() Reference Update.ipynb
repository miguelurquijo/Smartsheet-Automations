# Update cell vlaue with Todays Date

This scrip updates today reference with Today's date every day in a sheet that is used in multiple formuals across the Growth Paltform

```python
x = datetime.date.today()
todays_date = x.strftime('%m/%d/%Y')
start_time = time.time()


new_cell = smartsheet.models.Cell()
new_cell.column_id = 'COLUMN-ID  WHERE THE DATE WILL BE INPUTED'
new_cell.value = todays_date
new_cell.strict = False

# Build the row to update
new_row = smartsheet.models.Row()
new_row.id = 'ROW-ID WHERE THE TODAY REFERENCE WIL BE INPUTED'
new_row.cells.append(new_cell)

# Update rows
updated_row = smartsheet_client.Sheets.update_rows(
  1967977890375556,      # sheet_id
  [new_row])

print(todays_date)
print("--- %s seconds ---" % (time.time() - start_time))
```
